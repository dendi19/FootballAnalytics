```python
import pandas as pd
import datetime
from time import sleep
from tqdm import tqdm
```

```python
perfdf = pd.read_csv("combinedall.csv")
scoredf = pd.read_csv("...\\spreadspoke_scores.csv")
stadiumdf = pd.read_csv("...\\nfl_stadiums.csv")
```

```python
perfdf = perfdf.drop(['Unnamed: 0'], axis=1)
```

```python
perfdf
```

```python
# convert the 'Date' column to datetime format
scoredf['schedule_date']= pd.to_datetime(scoredf['schedule_date'])
```

```python
scoredf
```

```python
testdf = perfdf.loc[(perfdf['Year'] >= 2016) & (perfdf['Year'] < 2026) ]
testdf
```

```python
print(datetime.datetime.now())
for d,i in tqdm(testdf.iterrows()):
    try:
        stadium = scoredf['stadium'][(scoredf['team_home'] == f'{i[3]}') & (scoredf['schedule_date'] == f'{i[0]}')].item()
    except:
        stadium = scoredf['stadium'][(scoredf['team_away'] == f'{i[3]}') & (scoredf['schedule_date'] == f'{i[0]}')].item()
    print(i[2], stadium, lat, long, elev)
    lat = stadiumdf['LATITUDE'][(stadiumdf['stadium_name'] == stadium)].item()
    long = stadiumdf['LONGITUDE'][(stadiumdf['stadium_name'] == stadium)].item()
    elev = stadiumdf['ELEVATION'][(stadiumdf['stadium_name'] == stadium)].item()
    testdf['lat'][(testdf['Tm'] == f'{i[3]}') & (testdf['Date'] == f'{i[0]}')] = lat
    testdf['long'][(testdf['Tm'] == f'{i[3]}') & (testdf['Date'] == f'{i[0]}')] = long
    testdf['elevation'][(testdf['Tm'] == f'{i[3]}') & (testdf['Date'] == f'{i[0]}')] = elev
    print(i[2], stadium, lat, long, elev)
print(datetime.datetime.now())
```
