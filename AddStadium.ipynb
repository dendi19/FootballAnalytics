```python
scores['team_home'][(scores['team_home'] == 'Tennessee Titans') & (scores['schedule_date'] == '11/12/2020')].item()
```

```python
testdf = stadbox
testdf['home_away'] = 0
testdf['win'] = 0
testdf['temp']=0
testdf['wind']=0
testdf
```

```python
print(datetime.datetime.now())
for d,i in testdf.iterrows():
    try:
        winloss = scores['winner'][(scores['team_home'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        stadium = scores['team_home'][(scores['team_home'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        temp = scores['weather_temperature'][(scores['team_home'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        wind = scores['weather_wind_mph'][(scores['team_home'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        testdf['home_away'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = 'home'
        testdf['temp'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = temp
        testdf['wind'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = wind
        if winloss == 'home':
            testdf['win'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = 1
    except:
        winloss = scores['winner'][(scores['team_away'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        stadium = scores['team_away'][(scores['team_away'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        temp = scores['weather_temperature'][(scores['team_away'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        wind = scores['weather_wind_mph'][(scores['team_away'] == f'{i[4]}') & (scores['schedule_date'] == f'{i[1]}')].item()
        testdf['home_away'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = 'away'
        testdf['temp'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = temp
        testdf['wind'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = wind
        if winloss == 'away':
            testdf['win'][(testdf['Tm'] == f'{i[4]}') & (testdf['Date'] == f'{i[1]}')] = 1
print(datetime.datetime.now()) 
```
